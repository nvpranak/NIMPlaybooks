## Configure NGC using the cli

### Step 1: Download and install the CLI
You can follow these steps and setup an API key.


Installing the NGC CLI
Navigate to https://catalog.ngc.nvidia.com/orgs/nvidia/teams/ngc-apps/resources/ngc_cli.

Select Download.

Select one of the download methods.

If you select Browser (Direct Download) or WGET, the appropriate command is copied to the clipboard. Use that command to install the ngc cli.

If you select CLI, a zip file is downloaded. Unzip the contents of the zip file and add the appropriate executable to your path.

### Step 2: Configure an API key.

```
ngc config set
Enter API key [********************************************************************************MzFh]. Choices: [<VALID_APIKEY>, 'no-apikey']:
Enter org:
Enter team:
Enter ace [no-ace]:
Successfully saved NGC configuration to /home/<user>/.ngc/config
```


## Follow the instructions from the following reference

https://developer.nvidia.com/docs/nemo-microservices/embedding/source/deploy.html



#### Use the following commands to specify the environment variables for the examples.

```
export HOSTNAME="0.0.0.0"
export SERVICE_PORT=8080
export CONTAINER_NAME="embedding-ms"
```


#### Downloading the model
Use the following commands to download the NV-Embed-QA model from NGC.

```
ngc registry model download-version "ohlfw0olaadg/ea-participants/nv-embed-qa:4"
chmod -R o+rX nv-embed-qa_v4
MODEL_ID=NV-Embed-QA
MODEL_NAME=nv-embed-qa_v4/NV-Embed-QA-4.nemo
```

#### Launching the microservice container
Use the following command to start NeMo Retriever embedding within a docker container.

```
docker run \
  -it \
  --name $CONTAINER_NAME \
  --gpus=1 \
  --shm-size=8G \
  -v $(pwd):/model-checkpoint-path \
  -p $SERVICE_PORT:$SERVICE_PORT \
  nvcr.io/ohlfw0olaadg/ea-participants/nemo-retriever-embedding-microservice:24.02 \
  bin/web -p $SERVICE_PORT -c /model-checkpoint-path/$MODEL_NAME -g model_config_templates/${MODEL_ID}_template.yaml
```

### Running Inference
NOTE: It may take a few seconds for the microservice to be ready and start accepting requests from the time the docker container is started.

In the previous examples, the OpenAI compatible API endpoint is exposed on port $SERVICE_PORT on the host where that command is executed.

To run a sample request, set the HOSTNAME and MODEL parameters, as shown in the following example. Make sure the MODEL_ID variable is set according to the downloaded model as previously specified.

```
curl -X "POST" \
  "http://${HOSTNAME}:${SERVICE_PORT}/v1/embeddings" \
  -H 'accept: application/json' \
  -H 'Content-Type: application/json' \
  -d '{
    "input": ["Hello world"],
    "model": "'"$MODEL_ID"'",
    "input_type": "query"
}'
```
